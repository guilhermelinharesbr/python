Cores no Terminal na Prática

```python
print('Olá Mundo!')
```

```python
print('\033[31mOlá Mundo em vermelho!')
```

```python
print('\033[31;43mOlá Mundo em vermelho e com fundo amarelo!')
```

```python
print('\033[1;31;43mOlá Mundo em negrito, cor vermelha e com fundo amarelo!')
```

```python
print('\033[1;31;43mOlá Mundo em negrito, cor vermelha e com fundo amarelo e que fique a cor do fundo amarela apenas até o final desta frase e não pro terminal todo!\033[m')
```

```python
#Mostrando um número de uma cor e o outro de outra cor:
a = 3
b = 5
print('Os valores são \033[32m{}\033[0m e \033[31m{}\033[0m !!!'.format(a,b))
```

```python
#Mostrando uma forma diferente de usar as cores
nome = 'Guilherme'
print('Olá! Muito prazer em te conhecer, {}{}{}!!!'.format('\033[4;34m', nome, '\033[0m'))
```

```python
#Mostrando uma forma diferente de usar as cores
'limpa'
nome = 'Guilherme'
cores = {'limpa':'\033[0m',
         'azul':'\033[34m',
         'amarelo':'\033[33m', 
         'pretoebranco':'\033[7;30m'}
print('Olá! Muito prazer em te conhecer, {}{}{}!!!'.format(cores['amarelo'], nome, cores['limpa']))
```
